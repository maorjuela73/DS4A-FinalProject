```python
### Meses desde que se actualizo Last Updated
tabla_1['fecha']=tabla_1['Last Updated'].str.split(' ')
tabla_1['len']=tabla_1['fecha'].apply(lambda x: len(x))
tabla_1=tabla_1[tabla_1['len']>2]
tabla_1['mes']=tabla_1['fecha'].apply(lambda x: strptime(x[0][:3],'%b').tm_mon)
tabla_1['mes']=tabla_1['mes'].apply(lambda x: '0'+str(x) if len(str(x))<=1 else str(x))
tabla_1['dia']=tabla_1['fecha'].apply(lambda x: x[1].split(',')[0])
tabla_1['dia']=tabla_1['dia'].apply(lambda x: '0'+str(x) if len(str(x))<=1 else str(x))
tabla_1['anio']=tabla_1['fecha'].apply(lambda x: x[2])
tabla_1['fecha_1']=tabla_1['anio']+tabla_1['mes']+tabla_1['dia']
tabla_1['fecha_2']=pd.to_datetime(tabla_1['fecha_1'])
tabla_1['days_since_last_up']=(pd.to_datetime('2019-04-04 00:00:00')-tabla_1['fecha_2']).dt.days
```

```python
def get_ct(df, var):
    ct_res = pd.crosstab(df[var], df['success'], margins=True)
    ct_res["success(%)"] = round( ct_res[1]/ct_res["All"]*100, 2)
    ct_res.sort_values(by='success(%)',ascending=False,inplace=True)
    return ct_res.drop(columns="All")
```

```python
X=tabla_1[['Intercept','Size','Price_1','Latest_Version','days_since_last_up','tamano_nombre','Category_BOOKS_AND_REFERENCE','Category_BUSINESS','Category_COMMUNICATION','Category_EDUCATION','Category_ENTERTAINMENT','Category_FINANCE','Category_FOOD_AND_DRINK','Category_GAME','Category_HEALTH_AND_FITNESS','Category_LIFESTYLE','Category_MAPS_AND_NAVIGATION','Category_MEDICAL','Category_MUSIC_AND_AUDIO','Category_NEWS_AND_MAGAZINES','Category_Other','Category_PERSONALIZATION','Category_PHOTOGRAPHY','Category_PRODUCTIVITY','Category_SHOPPING','Category_SOCIAL','Category_SPORTS','Category_TOOLS','Content_Rating_Everyone_10','Content_Rating_Adults_only_18','Content_Rating_Teen','Content_Rating_Unrated','Minimum_Version_depends_on_device_0']]
imputer=Imputer(missing_values='NaN',strategy='mean',axis=0)
imputer=imputer.fit(X)
X=imputer.transform(X.values)
X=pd.DataFrame.from_records(X)
X.columns=['Intercept','Size','Price_1','Latest_Version','days_since_last_up','tamano_nombre','Category_BOOKS_AND_REFERENCE','Category_BUSINESS','Category_COMMUNICATION','Category_EDUCATION','Category_ENTERTAINMENT','Category_FINANCE','Category_FOOD_AND_DRINK','Category_GAME','Category_HEALTH_AND_FITNESS','Category_LIFESTYLE','Category_MAPS_AND_NAVIGATION','Category_MEDICAL','Category_MUSIC_AND_AUDIO','Category_NEWS_AND_MAGAZINES','Category_Other','Category_PERSONALIZATION','Category_PHOTOGRAPHY','Category_PRODUCTIVITY','Category_SHOPPING','Category_SOCIAL','Category_SPORTS','Category_TOOLS','Content_Rating_Everyone_10','Content_Rating_Adults_only_18','Content_Rating_Teen','Content_Rating_Unrated','Minimum_Version_depends_on_device_0']
X.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,tabla_1['success'],test_size=0.3,random_state=42)


```

```python
# code the discrete variable by the specification above
logit = sm.Logit(list(Y_train),X_train)
logit_res = logit.fit()
print( logit_res.summary())
```

```python
y_pred_train=logit_res.predict(X_train)
y_pred_test=logit_res.predict(X_test)
y_pron_train=np.where(y_pred_train>0.5,1,0)
y_pron_test=np.where(y_pred_test>0.5,1,0)
```

```python
fpr_train,tpr_train,thresholds_train=metrics.roc_curve(Y_train,y_pred_train)
fpr_test,tpr_test,thresholds_test=metrics.roc_curve(Y_test,y_pred_test)

print('AUC train'+ str(metrics.auc(fpr_train,tpr_train)))
print('AUC test'+ str(metrics.auc(fpr_test,tpr_test)))
```

### construccion scorecards

```python
scorecard=pd.DataFrame()
scorecard['probabilidad']=y_pred_train
scorecard['success']=Y_train
scorecard['decil']=pd.qcut(scorecard['probabilidad'],10,duplicates='drop')
scorecard_1=scorecard.groupby(['decil']).agg({'probabilidad':'count','success':'sum'}).reset_index()
scorecard_1['tasa_buenos']=(scorecard_1['success']/scorecard_1['probabilidad'])*100
scorecard_1['dist']=(scorecard_1['probabilidad']/scorecard_1['probabilidad'].sum())*100
scorecard_1.head(10)
```

### entre malos no discrimina pero si identifica muy bien a los exitosos
