### Getting filter Df by name

```python
appName = "WWE"
dffilter1 = df0[df0['App Name']==appName]
categoryTemp = dffilter1['Category'].tolist()[0]
print(categoryTemp)
print(sum(df0['Category'].isnull()))
```

```python
#dffilter2 = df0.dropna(subset=['Category'])
dffilter2 = df0[df0['Category']==categoryTemp]
dffilter2
```

```python
dfmerge = pd.merge(dffilter2,df2,on='App Name')
rowbyAppName = dfmerge[dfmerge['App Name'] == appName]
rowbyAppName
```

```python
dfmergefinal = dfmerge[dfmerge['App Name'] != appName]
dfmergefinalsort = dfmergefinal.sort_values(by=['ind_total_norm'],ascending=False)
dfmergefinalsort[['name','Price','tamano_nombre',
                  'Category_y','Content Rating_y',
                  'ind_total_norm','success']]
```
