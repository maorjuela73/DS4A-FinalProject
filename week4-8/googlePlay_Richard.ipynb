```python
###Limpia Installs
tabla_1['Installs']=np.where(tabla_1['Installs'].isin(['0+','1+','5+','10+','50+']),'0+',
                              np.where(tabla_1['Installs'].isin(['6','20','54','10','1','13','59','71',' Xmax X','EDUCATION','4.823529243']),'0+',
                                       np.where(tabla_1['Installs'].isin(['156','117','166']),'100+',
                                               np.where(tabla_1['Installs'].isin(['511','927']),'500+',
                                                       np.where(tabla_1['Installs'].isin(['11976']),'10,000+',tabla_1['Installs'])))))
                

```

```python
###Limpia size
tabla_1['size_depends_on_device']=np.where(tabla_1['Size']=='Varies with device',1,0)
tabla_1['Size']=np.where(tabla_1['Size'].str.find('M')>0,(pd.to_numeric(tabla_1['Size'].str.replace('M',''),errors='coerce'))*1000,
                           np.where(tabla_1['Size'].str.find('K')>0,pd.to_numeric(tabla_1['Size'].str.replace('K',''),errors='coerce'),np.nan))

```

```python
tabla_1['Content Rating']=np.where(tabla_1['Content Rating'].isin(['0','$0.99','100,000+','17M','$2.49','3702']),'Unrated',
                                     np.where(tabla_1['Content Rating']=='Mature 17+','Adults only 18+',tabla_1['Content Rating']))
```

```python
tabla_1['Minimum Version_depends_on_device']=np.where(tabla_1['Minimum Version']=='Varies with device',1,0)
tabla_1['Minimum Version']=pd.to_numeric(tabla_1['Minimum Version'].str[0:3],errors='coerce')
                            

```

```python
tabla_1['Latest Version_depends_on_device']=np.where(tabla_1['Latest Version']=='Varies with device',1,0)
tabla_1['Latest Version']=pd.to_numeric(tabla_1['Latest Version'].str[0:3],errors='coerce')
                       
```
